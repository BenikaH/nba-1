### Add salary data to dfs_player_gamelogs

Adds in salrank_team (1 is highest), salrank_slate, totsal (per team), salratio (player/team)

```python
%%sql
WITH psal AS (
  SELECT *,
    row_number() OVER (PARTITION BY dfs_player_gamelogs.season, dfs_player_gamelogs.game_id, dfs_player_gamelogs.team_code ORDER BY dfs_player_gamelogs.salary DESC) AS salrank_team,
    row_number() OVER (PARTITION BY dfs_player_gamelogs.season, dfs_player_gamelogs.game_date ORDER BY dfs_player_gamelogs.salary DESC) AS salrank_slate
  FROM dfs_player_gamelogs
), 
teamsal AS (
  SELECT 
    psal_1.season, psal_1.game_id, psal_1.game_date, psal_1.team_code,
    sum(psal_1.salary) AS totsal
  FROM psal psal_1
  WHERE psal_1.salrank_team < 9
  GROUP BY psal_1.season, psal_1.game_id, psal_1.game_date, psal_1.team_code
),
psal2 AS (
  SELECT 
    psal.*,
    teamsal.totsal,
    round(psal.salary / teamsal.totsal, 3) AS salratio
  FROM psal
  LEFT JOIN teamsal ON psal.game_id = teamsal.game_id AND psal.team_code::text = teamsal.team_code::text
)

SELECT * FROM psal2
ORDER BY psal2.season, psal2.team_code, psal2.game_date
LIMIT 10;
```

### Add missing player data to dfs_player_gamelogs

Adds in min_msg (combined avg minutes of missing players), fga_msg (same for fga), pts_msg (same for pts), dk_msg (same for dk_points).

```python
%%sql
WITH psal AS (
  SELECT *,
    row_number() OVER (PARTITION BY dfs_player_gamelogs.season, dfs_player_gamelogs.game_id, dfs_player_gamelogs.team_code ORDER BY dfs_player_gamelogs.salary DESC) AS salrank_team,
    row_number() OVER (PARTITION BY dfs_player_gamelogs.season, dfs_player_gamelogs.game_date ORDER BY dfs_player_gamelogs.salary DESC) AS salrank_slate
  FROM dfs_player_gamelogs
), 
teamsal AS (
  SELECT 
    psal_1.season, psal_1.game_id, psal_1.game_date, psal_1.team_code,
    sum(psal_1.salary) AS totsal
  FROM psal psal_1
  WHERE psal_1.salrank_team < 9
  GROUP BY psal_1.season, psal_1.game_id, psal_1.game_date, psal_1.team_code
),
psal2 AS (
  SELECT 
    psal.*,
    teamsal.totsal,
    round(psal.salary / teamsal.totsal, 3) AS salratio
  FROM psal
  LEFT JOIN teamsal ON psal.game_id = teamsal.game_id AND psal.team_code::text = teamsal.team_code::text
)

SELECT 
  gl.*,
  mps.min_msg, mps.fga_msg, mps.pts_msg, mps.dk_msg
FROM psal2 gl
LEFT JOIN missing_playerstats mps ON gl.game_id = mps.game_id AND gl.team_id = mps.team_id
ORDER BY game_id, nbacom_player_id
LIMIT 10
```

## Add data from previous games (what you know before game starts)

Add ceiling calculated as best game, average of best 3 games, and average of best 5 games.

```python
%%sql
WITH psal AS (
  SELECT *,
    row_number() OVER (PARTITION BY dfs_player_gamelogs.season, dfs_player_gamelogs.game_id, dfs_player_gamelogs.team_code ORDER BY dfs_player_gamelogs.salary DESC) AS salrank_team,
    row_number() OVER (PARTITION BY dfs_player_gamelogs.season, dfs_player_gamelogs.game_date ORDER BY dfs_player_gamelogs.salary DESC) AS salrank_slate
  FROM dfs_player_gamelogs
), 
teamsal AS (
  SELECT 
    psal_1.season, psal_1.game_id, psal_1.game_date, psal_1.team_code,
    sum(psal_1.salary) AS totsal
  FROM psal psal_1
  WHERE psal_1.salrank_team < 9
  GROUP BY psal_1.season, psal_1.game_id, psal_1.game_date, psal_1.team_code
),
psal2 AS (
  SELECT 
    psal.*,
    teamsal.totsal,
    round(psal.salary / teamsal.totsal, 3) AS salratio
  FROM psal
  LEFT JOIN teamsal ON psal.game_id = teamsal.game_id AND psal.team_code::text = teamsal.team_code::text
),
gl AS (
  SELECT 
    gl.*,
    mps.min_msg, mps.fga_msg, mps.pts_msg, mps.dk_msg
  FROM psal2 gl
  LEFT JOIN missing_playerstats mps ON gl.game_id = mps.game_id AND gl.team_id = mps.team_id
),
gl2 AS (
  SELECT 
    gl.*,
    sq.dkp
  FROM gl
  LEFT JOIN 
    (SELECT gl_1.game_date,
            gl_1.nbacom_player_id,
            sort_desc(array_agg(gl_1.dk_points::integer) OVER w) AS dkp
     FROM gl gl_1
     WINDOW w AS (PARTITION BY gl_1.season, gl_1.nbacom_player_id ORDER BY gl_1.game_date ROWS BETWEEN UNBOUNDED PRECEDING AND 1 PRECEDING)) sq 
  ON gl.game_date = sq.game_date AND gl.nbacom_player_id = sq.nbacom_player_id
)

SELECT
  gl2.*,
  round(array_avg(subarray(gl2.dkp, 1, 1)::double precision[])::numeric, 2) AS ceiling1,
  round(array_avg(subarray(gl2.dkp, 1, 3)::double precision[])::numeric, 2) AS ceiling3,
  round(array_avg(subarray(gl2.dkp, 1, 5)::double precision[])::numeric, 2) AS ceiling5
FROM gl2
ORDER BY gl2.game_date, gl2.nbacom_player_id
LIMIT 20
```

Add in exponential moving averages and odds

```python
%%sql
WITH psal AS (
  SELECT *,
    row_number() OVER (PARTITION BY dfs_player_gamelogs.season, dfs_player_gamelogs.game_id, dfs_player_gamelogs.team_code ORDER BY dfs_player_gamelogs.salary DESC) AS salrank_team,
    row_number() OVER (PARTITION BY dfs_player_gamelogs.season, dfs_player_gamelogs.game_date ORDER BY dfs_player_gamelogs.salary DESC) AS salrank_slate
  FROM dfs_player_gamelogs
), 
teamsal AS (
  SELECT 
    psal_1.season, psal_1.game_id, psal_1.game_date, psal_1.team_code,
    sum(psal_1.salary) AS totsal
  FROM psal psal_1
  WHERE psal_1.salrank_team < 9
  GROUP BY psal_1.season, psal_1.game_id, psal_1.game_date, psal_1.team_code
),
psal2 AS (
  SELECT 
    psal.*,
    teamsal.totsal,
    round(psal.salary / teamsal.totsal, 3) AS salratio
  FROM psal
  LEFT JOIN teamsal ON psal.game_id = teamsal.game_id AND psal.team_code::text = teamsal.team_code::text
),
gl AS (
  SELECT 
    gl.*,
    mps.min_msg, mps.fga_msg, mps.pts_msg, mps.dk_msg
  FROM psal2 gl
  LEFT JOIN missing_playerstats mps ON gl.game_id = mps.game_id AND gl.team_id = mps.team_id
),
gl2 AS (
  SELECT 
    gl.*,
    sq.dkp
  FROM gl
  LEFT JOIN 
    (SELECT gl_1.game_date,
            gl_1.nbacom_player_id,
            sort_desc(array_agg(gl_1.dk_points::integer) OVER w) AS dkp
     FROM gl gl_1
     WINDOW w AS (PARTITION BY gl_1.season, gl_1.nbacom_player_id ORDER BY gl_1.game_date ROWS BETWEEN UNBOUNDED PRECEDING AND 1 PRECEDING)) sq 
  ON gl.game_date = sq.game_date AND gl.nbacom_player_id = sq.nbacom_player_id
),
gl3 AS (
  SELECT
    gl2.*,
    round(array_avg(subarray(gl2.dkp, 1, 1)::double precision[])::numeric, 2) AS ceiling1,
    round(array_avg(subarray(gl2.dkp, 1, 3)::double precision[])::numeric, 2) AS ceiling3,
    round(array_avg(subarray(gl2.dkp, 1, 5)::double precision[])::numeric, 2) AS ceiling5
  FROM gl2
)

SELECT
  gl3.*,
  round(avg(gl3.min) OVER w, 1) AS minavg,
  round(ema(gl3.min::numeric, 0.02) OVER w, 1) AS minema2,
  round(ema(gl3.min::numeric, 0.1) OVER w, 1) AS minema10,
  round(ema(gl3.min::numeric, 0.25) OVER w, 1) AS minema25,
  round(ema(gl3.min::numeric, 0.4) OVER w, 1) AS minema40,
  round(avg(gl3.dk_points) OVER w, 2) AS dkavg,
  round(ema(gl3.dk_points, 0.02) OVER w, 2) AS dkema2,
  round(ema(gl3.dk_points, 0.1) OVER w, 2) AS dkema10,
  round(ema(gl3.dk_points, 0.25) OVER w, 2) AS dkema25,
  round(ema(gl3.dk_points, 0.4) OVER w, 2) AS dkema40,
  COALESCE(gl3.consensus_spread, gl3.opening_spread) AS spread,
  COALESCE(gl3.consensus_game_ou, gl3.opening_game_ou) AS game_ou,
  COALESCE(gl3.consensus_implied_total, gl3.opening_implied_total) AS impltot
FROM
  gl3
WINDOW w AS (PARTITION BY season, nbacom_player_id ORDER BY game_date ROWS BETWEEN UNBOUNDED PRECEDING AND 1 PRECEDING)
ORDER BY
  season, game_date, nbacom_player_id
LIMIT 10
```

## Add in player careeer / bio

```python
%%sql
CREATE TEMP TABLE gl5 AS
WITH psal AS (
  SELECT *,
    row_number() OVER (PARTITION BY dfs_player_gamelogs.season, dfs_player_gamelogs.game_id, dfs_player_gamelogs.team_code ORDER BY dfs_player_gamelogs.salary DESC) AS salrank_team,
    row_number() OVER (PARTITION BY dfs_player_gamelogs.season, dfs_player_gamelogs.game_date ORDER BY dfs_player_gamelogs.salary DESC) AS salrank_slate
  FROM dfs_player_gamelogs
), 
teamsal AS (
  SELECT 
    psal_1.season, psal_1.game_id, psal_1.game_date, psal_1.team_code,
    sum(psal_1.salary) AS totsal
  FROM psal psal_1
  WHERE psal_1.salrank_team < 9
  GROUP BY psal_1.season, psal_1.game_id, psal_1.game_date, psal_1.team_code
),
psal2 AS (
  SELECT 
    psal.*,
    teamsal.totsal,
    round(psal.salary / teamsal.totsal, 3) AS salratio
  FROM psal
  LEFT JOIN teamsal ON psal.game_id = teamsal.game_id AND psal.team_code::text = teamsal.team_code::text
),
gl AS (
  SELECT 
    gl.*,
    mps.min_msg, mps.fga_msg, mps.pts_msg, mps.dk_msg
  FROM psal2 gl
  LEFT JOIN missing_playerstats mps ON gl.game_id = mps.game_id AND gl.team_id = mps.team_id
),
gl2 AS (
  SELECT 
    gl.*,
    sq.dkp
  FROM gl
  LEFT JOIN 
    (SELECT gl_1.game_date,
            gl_1.nbacom_player_id,
            sort_desc(array_agg(gl_1.dk_points::integer) OVER w) AS dkp
     FROM gl gl_1
     WINDOW w AS (PARTITION BY gl_1.season, gl_1.nbacom_player_id ORDER BY gl_1.game_date ROWS BETWEEN UNBOUNDED PRECEDING AND 1 PRECEDING)) sq 
  ON gl.game_date = sq.game_date AND gl.nbacom_player_id = sq.nbacom_player_id
),
gl3 AS (
  SELECT
    gl2.*,
    round(array_avg(subarray(gl2.dkp, 1, 1)::double precision[])::numeric, 2) AS ceiling1,
    round(array_avg(subarray(gl2.dkp, 1, 3)::double precision[])::numeric, 2) AS ceiling3,
    round(array_avg(subarray(gl2.dkp, 1, 5)::double precision[])::numeric, 2) AS ceiling5
  FROM gl2
),
gl4 AS (
SELECT
  gl3.*,
  round(avg(gl3.min) OVER w, 1) AS minavg,
  round(ema(gl3.min::numeric, 0.02) OVER w, 1) AS minema2,
  round(ema(gl3.min::numeric, 0.1) OVER w, 1) AS minema10,
  round(ema(gl3.min::numeric, 0.25) OVER w, 1) AS minema25,
  round(ema(gl3.min::numeric, 0.4) OVER w, 1) AS minema40,
  round(avg(gl3.dk_points) OVER w, 2) AS dkavg,
  round(ema(gl3.dk_points, 0.02) OVER w, 2) AS dkema2,
  round(ema(gl3.dk_points, 0.1) OVER w, 2) AS dkema10,
  round(ema(gl3.dk_points, 0.25) OVER w, 2) AS dkema25,
  round(ema(gl3.dk_points, 0.4) OVER w, 2) AS dkema40,
  COALESCE(gl3.consensus_spread, gl3.opening_spread) AS spread,
  COALESCE(gl3.consensus_game_ou, gl3.opening_game_ou) AS game_ou,
  COALESCE(gl3.consensus_implied_total, gl3.opening_implied_total) AS impltot
FROM
  gl3
WINDOW w AS (PARTITION BY season, nbacom_player_id ORDER BY game_date ROWS BETWEEN UNBOUNDED PRECEDING AND 1 PRECEDING)
)

SELECT
  gl4.*, (p.draft_round -1 + p.draft_number) as picknum, p.draft_round, COALESCE(ROUND((gl4.game_date - p.birthdate)::numeric/365,2), NULL) AS age
FROM
  gl4
LEFT JOIN players p ON gl4.nbacom_player_id = p.nbacom_player_id;

ALTER TABLE gl5 DROP COLUMN dkp;

SELECT * FROM gl5
ORDER BY random()
LIMIT 10;
```
