# NBA Depth Charts

Right now, if starters miss game and come back, then subsequent game is as if starters are missing even though starters are actually back. Want to make the 'starters_missing' metric more accurate.

One idea is to have a depth_charts table, which is the team's depth chart on every gameday. The starting place for the depth chart is the opening night lineup.

Will need some methodology for dealing with longer-term injuries. Maybe a 3-5 game window after important player is injured that I would change the depth chart because at that point minutes, production and salary have likely adjusted to absence.

## Step Two: Identify Thresholds for Starter

Can also count the number of starts prior to the game. Can designate player as regular starter if he has started more than X% of the team's game up to that point. To capture players coming off the injured list who haven't started a certain % of games, can look at their minutes per game and figure out if it is consistent with a starter.


Idea here is that some players are always starters, such as LeBron James. We can mark that player based on certain thresholds, such as minutes per game, points per game, etc.  Will assume that starters at the beginning of the year remain starters unless there is evidence to the contrary.

```python
%%sql
WITH t AS (
  SELECT 
    s.season, s.game_id, s.game_date, s.team_id, c.team_code as tm, gm.game_number as gnum,
    s.nbacom_player_id as pid, s.player_name, 
    s.is_starter, s.start_position as stpos,
    COALESCE(sum(CASE WHEN s.is_starter THEN 1 ELSE 0 END) over w,0) AS num_starts,
    ROUND(lag(min_played,1) over w / lag(gp, 1) over w, 2)AS minpg, 
    ROUND(lag(pts,1) over w / lag(gp, 1) over w, 2) AS ppg
  FROM
    starters s
  JOIN playerstats_daily ps ON s.game_date = ps.as_of AND s.nbacom_player_id = ps.nbacom_player_id
  JOIN cteams c ON s.team_id = c.nbacom_team_id
  JOIN gamesmeta gm ON s.game_id = gm.game_id AND c.team_code = gm.team_code
  WHERE s.season = 2017
  WINDOW w AS (PARTITION BY s.season, s.nbacom_player_id ORDER BY s.game_id ROWS BETWEEN UNBOUNDED PRECEDING AND 1 PRECEDING)
)

SELECT
  t.*, (CASE WHEN gnum > 4 AND minpg >= 30 THEN TRUE 
        WHEN gnum <= 3 AND is_starter THEN TRUE
        WHEN num_starts/gnum::numeric > .5 THEN TRUE
        ELSE FALSE END) AS reg_starter
FROM t
ORDER BY season, tm, game_date, stpos
LIMIT 25
```

### Join other tables: gamesmeta and cteams

We have to join cteams to get the team_code (from the team_id) which, in turn, will allow a join on the gamesmeta table which will give us the game_number. 

```python
%%sql
WITH t AS (
  SELECT 
    s.season, s.game_id, s.game_date, s.team_id, c.team_code as tm, gm.game_number as gnum,
    s.nbacom_player_id as pid, s.player_name as player, 
    s.is_starter, s.start_position as stpos,
    COALESCE(sum(CASE WHEN s.is_starter THEN 1 ELSE 0 END) over w,0) AS nstart,
    ROUND(lag(min_played,1) over w / lag(gp, 1) over w, 2)AS minpg, 
    ROUND(lag(pts,1) over w / lag(gp, 1) over w, 2) AS ppg
  FROM
    starters s
  JOIN playerstats_daily ps ON s.game_date = ps.as_of AND s.nbacom_player_id = ps.nbacom_player_id
  JOIN cteams c ON s.team_id = c.nbacom_team_id
  JOIN gamesmeta gm ON s.game_id = gm.game_id AND c.team_code = gm.team_code
  WHERE s.season = 2017
  WINDOW w AS (PARTITION BY s.season, s.nbacom_player_id ORDER BY s.game_id ROWS BETWEEN UNBOUNDED PRECEDING AND 1 PRECEDING)
)

SELECT * FROM t 
ORDER BY season, team_id, game_date, stpos
LIMIT 20;
```

### Now add the criteria for "regular starters"

```python
%%sql
WITH t AS (
  SELECT 
    s.season, s.game_id, s.game_date, s.team_id, c.team_code as tm, gm.game_number as gnum,
    s.nbacom_player_id as pid, s.player_name as player, 
    s.is_starter, s.start_position as stpos,
    COALESCE(sum(CASE WHEN s.is_starter THEN 1 ELSE 0 END) over w,0) AS nstart,
    ROUND(lag(min_played,1) over w / lag(gp, 1) over w, 2)AS minpg, 
    ROUND(lag(pts,1) over w / lag(gp, 1) over w, 2) AS ppg
  FROM
    starters s
  JOIN playerstats_daily ps ON s.game_date = ps.as_of AND s.nbacom_player_id = ps.nbacom_player_id
  JOIN cteams c ON s.team_id = c.nbacom_team_id
  JOIN gamesmeta gm ON s.game_id = gm.game_id AND c.team_code = gm.team_code
  WHERE s.season = 2017
  WINDOW w AS (PARTITION BY s.season, s.nbacom_player_id ORDER BY s.game_id ROWS BETWEEN UNBOUNDED PRECEDING AND 1 PRECEDING)
)

SELECT
  t.*, (CASE WHEN gnum > 4 AND minpg >= 30 THEN TRUE 
        WHEN gnum <= 3 AND is_starter THEN TRUE
        WHEN nstart/gnum::numeric > .5 THEN TRUE
        ELSE FALSE END) AS reg_starter
FROM t
ORDER BY season, tm, game_date, stpos
LIMIT 25;
```

### Sum up to see who is a regular starter

```python
%%sql
WITH t AS (
  SELECT 
    s.season, s.game_id, s.game_date, s.team_id, c.team_code as tm, gm.game_number as gnum,
    s.nbacom_player_id as pid, s.player_name as player, 
    s.is_starter, s.start_position as stpos,
    COALESCE(sum(CASE WHEN s.is_starter = 't' THEN 1 ELSE 0 END) over w,0) AS nstart,
    ROUND(lag(min_played,1) over w / lag(gp, 1) over w, 2)AS minpg, 
    ROUND(lag(pts,1) over w / lag(gp, 1) over w, 2) AS ppg
  FROM
    starters s
  JOIN playerstats_daily ps ON s.game_date = ps.as_of AND s.nbacom_player_id = ps.nbacom_player_id
  JOIN cteams c ON s.team_id = c.nbacom_team_id
  JOIN gamesmeta gm ON s.game_id = gm.game_id AND c.team_code = gm.team_code
  WHERE s.season = 2017
  WINDOW w AS (PARTITION BY s.season, s.nbacom_player_id ORDER BY s.game_id ROWS BETWEEN UNBOUNDED PRECEDING AND 1 PRECEDING)
),

t2 AS (
  SELECT
    t.*, (CASE WHEN gnum > 4 AND minpg >= 30 THEN 1 
          WHEN gnum <= 7 AND is_starter = 't' THEN 1
          WHEN gnum > 7 AND nstart/gnum::numeric > .75 THEN 1
          ELSE 0 END) AS reg_starter
  FROM t
),

t3 AS (
  SELECT 
    DISTINCT pid, player, 
    sum(reg_starter) over w AS nreg
  FROM t2
  WINDOW w AS (PARTITION BY season, pid)
)

SELECT t2.*, t3.nreg 
FROM t2
JOIN t3 ON t2.pid = t3.pid
ORDER BY season, tm, gnum, pid
LIMIT 100
```
